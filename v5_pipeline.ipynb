````python
### DAW Pipeline: Velo/Fuss + MeteoSwiss
import os
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

sns.set_style("whitegrid")

# ---------------- Configuration ----------------
PROJECT_PATH = Path("/Users/FHNW/Documents/daw/project")  # adapt if needed

try:
    os.chdir(PROJECT_PATH)
    print(f"Working directory: {os.getcwd()}")
except FileNotFoundError:
    raise FileNotFoundError(f"Path '{PROJECT_PATH}' not found!")

VELO_CSV = "converted_Velo_Fuss_Count.csv"
METEO_CSV = "ogd-smn_bas_h_historical_2020-2029.csv"
METEO_META = "ogd-smn_meta_parameters.csv"

YEARS = [2024]      # year(s) to analyze
CHUNKSIZE = 500_000 # rows per chunk when streaming Velo data

print("\nAnalyzing years:", YEARS, "with chunk size:", CHUNKSIZE, "\n")


# ---------------- MeteoSwiss: load + aggregate ----------------
def load_meteo_data(meteo_csv, meta_csv):
    """Load MeteoSwiss data and rename columns using metadata."""
    print("\n---------------------------------------------")
    print("\nLoading MeteoSwiss data and metadata...\n")
    data = pd.read_csv(meteo_csv, sep=";", encoding="latin1")
    meta = pd.read_csv(meta_csv, sep=";", encoding="latin1")

    print("Original MeteoSwiss data (head):")
    display(data.head(3))
    print("Metadata (head):")
    display(meta.head(3))

    rename_dict = dict(
        zip(meta["parameter_shortname"], meta["parameter_description_en"])
    )
    data = data.rename(columns=rename_dict)

    data["reference_timestamp"] = pd.to_datetime(
        data["reference_timestamp"],
        format="%d.%m.%Y %H:%M",
        dayfirst=True,
        errors="coerce",
    )

    print("\nMeteoSwiss data with parsed timestamps and readable column names:")
    display(data.head(3))
    return data


def aggregate_daily_meteo(df, years):
    """Aggregate hourly MeteoSwiss data to daily temp/precip/wind."""
    print("\nAggregating MeteoSwiss data to daily level...")
    df = df[df["reference_timestamp"].dt.year.isin(years)].copy()

    daily = (
        df.groupby(df["reference_timestamp"].dt.floor("D"))
        .agg(
            temp_mean_C=("Air temperature 2 m above ground; hourly mean", "mean"),
            precip_mm=("Precipitation; hourly total", "sum"),
            wind_mean_ms=("Wind speed scalar; hourly mean in m/s", "mean"),
        )
        .reset_index()
        .rename(columns={"reference_timestamp": "date"})
    )

    print("Daily MeteoSwiss data (head):")
    display(daily.head(3))
    return daily


# ---------------- Velo/Fuss: streaming + aggregate ----------------
def aggregate_velo_streaming(velo_csv, years, chunksize):
    """
    Stream Velo/Fuss CSV and aggregate to daily totals per site.

    Returns columns: ['SiteCode', 'SiteName', 'date', 'daily_total'].
    """
    print("\n---------------------------------------------")
    print("\nLoading Velo/Fuss data in streaming mode...")
    print("Aggregating daily totals while streaming...\n")

    usecols = ["Date", "SiteCode", "SiteName", "Total"]
    daily_agg = []
    chunk_count = 0

    for chunk in pd.read_csv(
        velo_csv,
        sep=";",
        encoding="utf-8",  # Basel portal is UTF-8; fixes 'Dreirosenbrücke'
        usecols=usecols,
        chunksize=chunksize,
        low_memory=False,
    ):
        chunk_count += 1

        # Parse dates and keep only valid ones in the selected years
        chunk["Date"] = pd.to_datetime(
            chunk["Date"], errors="coerce", dayfirst=True
        )
        chunk = chunk.dropna(subset=["Date"])
        chunk = chunk[chunk["Date"].dt.year.isin(years)]
        if chunk.empty:
            continue

        # Ensure numeric counts and drop negatives
        chunk["Total"] = pd.to_numeric(chunk["Total"], errors="coerce")
        chunk = chunk.dropna(subset=["Total"])
        chunk = chunk[chunk["Total"] >= 0]

        # Create daily date column and aggregate per site + date
        chunk["date"] = chunk["Date"].dt.floor("D")
        daily = (
            chunk.groupby(["SiteCode", "SiteName", "date"], as_index=False)["Total"]
            .sum()
        )
        daily_agg.append(daily)

    if not daily_agg:
        raise ValueError("No Velo/Fuss data found for requested years.")

    result = pd.concat(daily_agg, ignore_index=True)
    result = result.rename(columns={"Total": "daily_total"})

    # Second aggregation to ensure one row per (SiteCode, SiteName, date)
    result = (
        result.groupby(["SiteCode", "SiteName", "date"], as_index=False)["daily_total"]
        .sum()
    )


    print(f"Finished processing {chunk_count} chunks.")
    print("Daily Velo/Fuss data (head):")
    display(result.head(3))
    return result


# ---------------- Helper: missing days per year ----------------
def check_missing_days(df, col, years):
    """Count missing calendar days per year based on a datetime column."""
    out = {}
    for y in years:
        expected = pd.date_range(f"{y}-01-01", f"{y}-12-31", freq="D")
        present = df.loc[df[col].dt.year == y, col].drop_duplicates()
        missing = expected.difference(present)
        out[y] = len(missing)
    return out


# ---------------- Merge + simple feature engineering ----------------
def merge_datasets(daily_velo, daily_meteo):
    """
    Merge Velo/Fuss and Meteo data on 'date' and add weekday/weekend features.
    """
    print("\n---------------------------------------------")
    print("\nMerging Velo/Fuss and Meteo datasets...")
    daily_velo["date"] = pd.to_datetime(daily_velo["date"])
    daily_meteo["date"] = pd.to_datetime(daily_meteo["date"])

    merged = pd.merge(daily_velo, daily_meteo, on="date", how="left")

    merged["weekday"] = merged["date"].dt.day_name()
    merged["is_weekend"] = merged["date"].dt.dayofweek >= 5

    print(f"Merged dataset shape: {merged.shape}")
    print("\nMerged dataset (head):")
    display(merged.head(3))
    return merged


# ---------------- Plots (time series + scatter) ----------------
def plot_weekly_bar(merged_df, years):
    """Weekly total counts across all sites."""
    print("\n---------------------------------------------")
    print("\n\nGenerating weekly bar charts...")
    for year in years:
        df_year = merged_df[merged_df["date"].dt.year == year]
        weekly = (
            df_year.groupby(df_year["date"].dt.to_period("W"))["daily_total"]
            .sum()
            .reset_index()
        )
        weekly["week_start"] = weekly["date"].dt.start_time
        tick_idx = np.arange(0, len(weekly), 4)

        plt.figure(figsize=(12, 4))
        sns.barplot(data=weekly, x="week_start", y="daily_total")
        plt.title(f"Weekly total Velo/Fuss counts ({year})")
        plt.xlabel("Week start date")
        plt.ylabel("Total counts")
        plt.xticks(
            tick_idx,
            weekly["week_start"].dt.strftime("%Y-%m-%d").iloc[tick_idx],
            rotation=45,
        )
        plt.tight_layout()
        plt.show()


def plot_scatter_trends(merged_df, cols):
    """Scatterplots of daily_total vs selected meteo variables + linear trend."""
    print("\nGenerating scatterplots with linear trends...")
    for col in cols:
        df = merged_df[[col, "daily_total"]].dropna()
        x = df[col].values
        y = df["daily_total"].values

        coeffs = np.polyfit(x, y, 1)
        linear_func = np.poly1d(coeffs)
        x_sorted = np.sort(x)

        plt.figure(figsize=(8, 4))
        sns.scatterplot(x=x, y=y, alpha=0.4)
        plt.plot(x_sorted, linear_func(x_sorted), color="red", linewidth=1)
        plt.title(f"Daily total Velo/Fuss vs {col}")
        plt.xlabel(col)
        plt.ylabel("Daily total")
        plt.tight_layout()
        plt.show()


# ---------------- Some extra analysis ----------------
def analyze_relationships(merged_df):
    """
    Print and visualize simple group statistics to answer high-level research questions.

    - Temperature bins vs mean counts
    - Precipitation categories vs mean counts
    - Weekday vs weekend differences
    """
    df = merged_df.copy()

    # --- Temperature bins ---
    temp_bins = [-20, 0, 10, 15, 20, 25, 30, 45]
    temp_labels = ["<0", "0–10", "10–15", "15–20", "20–25", "25–30", ">30"]
    df["temp_bin"] = pd.cut(
        df["temp_mean_C"],
        bins=temp_bins,
        labels=temp_labels,
        include_lowest=True,
    )

    temp_means = df.groupby("temp_bin", observed=True)["daily_total"].mean().round(1)
    print("\nMean daily_total by temperature bin:")
    print(temp_means)

    temp_means_df = temp_means.reset_index(name="mean_daily_total")

    plt.figure(figsize=(8, 4))
    sns.barplot(data=temp_means_df, x="temp_bin", y="mean_daily_total")
    plt.ylabel("Mean daily total")
    plt.xlabel("Daily mean temperature (°C bin)")
    plt.title("Mean daily counts by temperature bin")
    plt.tight_layout()
    plt.show()

    # --- Precipitation categories ---
    precip_bins = [-0.01, 0.1, 5, 1000]
    precip_labels = ["Dry", "Light rain", "Rainy"]
    df["precip_cat"] = pd.cut(
        df["precip_mm"],
        bins=precip_bins,
        labels=precip_labels,
        include_lowest=True,
    )

    precip_means = df.groupby("precip_cat", observed=True)["daily_total"].mean().round(1)
    print("\nMean daily_total by precipitation category:")
    print(precip_means)

    precip_means_df = precip_means.reset_index(name="mean_daily_total")

    plt.figure(figsize=(8, 4))
    sns.barplot(data=precip_means_df, x="precip_cat", y="mean_daily_total")
    plt.ylabel("Mean daily total")
    plt.xlabel("Precipitation category")
    plt.title("Mean daily counts by precipitation category")
    plt.tight_layout()
    plt.show()

    # --- Weekday pattern ---
    weekday_means = df.groupby("weekday")["daily_total"].mean().round(1)

    order = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday",]

    print("\nMean daily_total by weekday:")
    print(weekday_means.reindex(order))  # <- imprime en orden Monday–Sunday

    weekday_means_df = weekday_means.reindex(order).reset_index(name="mean_daily_total")

    plt.figure(figsize=(8, 4))
    sns.barplot(
        data=weekday_means_df,
        x="weekday",
        y="mean_daily_total",
    )
    plt.ylabel("Mean daily total")
    plt.xlabel("Weekday")
    plt.title("Mean daily counts by weekday")
    plt.tight_layout()
    plt.show()

    # --- Weekday vs weekend ---
    weekend_means = df.groupby("is_weekend")["daily_total"].mean().round(1)
    print("\nMean daily_total: weekday vs weekend:")
    print(weekend_means)

    weekend_means_df = weekend_means.reset_index(name="mean_daily_total")
    weekend_means_df["label"] = weekend_means_df["is_weekend"].map(
        {False: "Weekday", True: "Weekend"}
    )

    plt.figure(figsize=(5, 4))
    sns.barplot(
        data=weekend_means_df,
        x="label",
        y="mean_daily_total",
    )
    plt.ylabel("Mean daily total")
    plt.xlabel("")
    plt.title("Mean daily counts: weekday vs weekend")
    plt.tight_layout()
    plt.show()


# ---------------- Correlation analysis ----------------
def analyze_correlations(merged_df):
    """
    Compute and visualize Pearson correlations between daily_total and numeric features.

    Includes:
    - Correlation matrix heatmap (Pearson)
    - Sorted absolute correlations with daily_total
    """
    print("\n---------------------------------------------")
    print("\n\nCorrelation analysis (daily_total vs numeric features)")
    df = merged_df.copy()

    # Select numeric columns of interest
    numeric_cols = ["daily_total", "temp_mean_C", "precip_mm", "wind_mean_ms"]
    numeric_cols = [c for c in numeric_cols if c in df.columns]

    if len(numeric_cols) < 2:
        print("Not enough numeric columns available for correlation analysis.")
        return

    corr_df = df[numeric_cols].corr(method="pearson")

    # Upper-triangular mask for a cleaner heatmap
    mask = np.triu(np.ones_like(corr_df, dtype=bool))

    plt.figure(figsize=(6, 4))
    sns.heatmap(
        corr_df,
        mask=mask,
        annot=True,
        fmt=".2f",
        cmap="coolwarm",
        center=0,
        vmin=-1,
        vmax=1,
        square=True,
        cbar_kws={"label": "Pearson correlation", "shrink": 0.7},
    )
    plt.title("Pearson correlation matrix (daily_total and weather features)")
    plt.tight_layout()
    plt.show()

    # Absolute correlations with daily_total
    if "daily_total" in corr_df.columns:
        corr_with_target = corr_df["daily_total"].drop("daily_total").abs().sort_values(ascending=False)
        print("\nTop absolute correlations with daily_total:")
        display(corr_with_target.to_frame("abs_corr"))


# ---------------- Run full pipeline ----------------
def run_pipeline():
    """Execute the full DAW pipeline and save merged dataset."""
    meteo_raw = load_meteo_data(METEO_CSV, METEO_META)
    daily_meteo = aggregate_daily_meteo(meteo_raw, YEARS)

    daily_velo = aggregate_velo_streaming(VELO_CSV, YEARS, CHUNKSIZE)

    # --- Simple data quality checks ---
    print("\nChecking missing days in both datasets:")
    print("MeteoSwiss:", check_missing_days(daily_meteo, "date", YEARS))
    print("Velo/Fuss:", check_missing_days(daily_velo, "date", YEARS))

    daily_meteo = aggregate_daily_meteo(meteo_raw, YEARS)

    # Check for duplicates in daily Velo/Fuss
    dup_count = daily_velo.duplicated(["SiteCode", "SiteName", "date"]).sum()
    print("\nPossible duplicates in daily Velo/Fuss (SiteCode, SiteName, date):", dup_count)

    merged = merge_datasets(daily_velo, daily_meteo)

    # Check missing values in merged dataset
    print("\nMissing values per column in merged dataset:")
    print(merged.isna().sum())

    plot_weekly_bar(merged, YEARS)
    plot_scatter_trends(merged, ["temp_mean_C", "precip_mm", "wind_mean_ms"])
    analyze_relationships(merged)
    analyze_correlations(merged)

    merged.to_csv("merged_dataset.csv", index=False)
    print("\nMerged dataset saved as 'merged_dataset.csv'.")
    return merged


# Run pipeline
merged_df = run_pipeline()
````

---

### MeteoSwiss: loading and renaming columns

* **What**
  Load MeteoSwiss hourly station data and its metadata; convert short cryptic column names into readable English descriptions and parse timestamps.

* **How**

  * Read both CSV files with `sep=";"` and `encoding="latin1"`.
  * Build `rename_dict` via `zip(meta["parameter_shortname"], meta["parameter_description_en"])`.
  * Rename columns and parse `reference_timestamp` as a proper `datetime`.

* **Why**

  * Using the official metadata to rename fields makes downstream analysis much clearer (e.g. `"Air temperature 2 m above ground; hourly mean"` instead of `tre200h0`).
  * Explicit date parsing with a known format and `errors="coerce"` ensures robust handling of malformed timestamps.
  * Having a clean, well-labeled dataframe is essential for reproducible wrangling.

---

### MeteoSwiss: hourly → daily aggregation

* **What**
  Aggregate hourly meteorological variables to daily summaries for the selected year(s).

* **How**

  * Filter by year: `df[df["reference_timestamp"].dt.year.isin(years)]`.
  * Group by `reference_timestamp.floor("D")`.
  * Compute:

    * `temp_mean_C`: mean daily air temperature (2 m above ground).
    * `precip_mm`: total daily precipitation.
    * `wind_mean_ms`: mean daily wind speed.
  * Rename the datetime column to `date`.

* **Why**

  * Traffic data is aggregated at the **daily** level; weather needs to match this resolution.
  * Mean temperature and wind, plus total precipitation, are standard daily climate indicators.
  * Aligning on a single `date` column provides a natural join key for later merging.

---

### Velo/Fuss: streaming, cleaning and daily aggregation (with second aggregation)

* **What**
  Stream the large Velo/Fuss dataset in chunks, clean it, and aggregate to daily totals per site.
  Guarantee exactly **one row per (SiteCode, SiteName, date)** by a second global aggregation.

* **How**

  * Use `pd.read_csv(..., chunksize=CHUNKSIZE, encoding="utf-8")` and restrict to needed columns.
  * For each chunk:

    * Parse `Date` with `errors="coerce"` and filter invalid dates.
    * Keep only rows from the selected years.
    * Convert `Total` to numeric, drop non-numeric and negative values.
    * Create `date = Date.dt.floor("D")`.
    * Group by `["SiteCode", "SiteName", "date"]` and sum `Total`.
  * Concatenate all partial results.
  * **Second aggregation step:**

    ```python
    result = (
        result.groupby(["SiteCode", "SiteName", "date"], as_index=False)["daily_total"]
        .sum()
    )
    ```

* **Why**

  * Chunked reading avoids loading the entire file into memory and scales to larger datasets.
  * Cleaning ensures we only keep valid dates and non-negative counts, which are meaningful for traffic analysis.
  * The **second aggregation** is crucial: when reading in chunks, rows for the same site-day can be split across chunks, so a final global `groupby` guarantees a clean data product where **each row represents the full daily total for one site**.

---

### Helper: checking missing days per year

* **What**
  Check calendar completeness: how many days are missing for each year.

* **How**

  * For each year in `YEARS`:

    * Build a full `pd.date_range` from `YYYY-01-01` to `YYYY-12-31`.
    * Compare against the set of present dates `df[col].dt.year == y`.
    * Count the difference.

* **Why**

  * Time series analysis strongly relies on understanding data coverage.
  * Verifying that there are no missing days (in this case 0 for both MeteoSwiss and Velo/Fuss in 2024) gives confidence that the merged dataset has full daily coverage without gaps.

---

### Merge and simple feature engineering

* **What**
  Merge daily traffic and daily weather on `date` and add calendar features (`weekday`, `is_weekend`).

* **How**

  * Convert both `daily_velo["date"]` and `daily_meteo["date"]` to `datetime`.
  * Use a **left join**:

    ```python
    merged = pd.merge(daily_velo, daily_meteo, on="date", how="left")
    ```
  * Add:

    * `weekday = merged["date"].dt.day_name()`
    * `is_weekend = merged["date"].dt.dayofweek >= 5`

* **Why**

  * Velo/Fuss is the **primary dataset** (object of interest); weather adds context → left join keeps all traffic rows.
  * Weekday and weekend flags capture strong calendar effects (commuting vs leisure), which are critical to interpret the impact of weather on traffic volumes.
  * The resulting merged dataset is a clean base for all subsequent plots and statistics.

---

### Visualisation: weekly totals (time series perspective)

* **What**
  Weekly bar plots of total counts across all sites, for each year in `YEARS`.

* **How**

  * Filter by year.
  * Group by `date.to_period("W")` and sum `daily_total`.
  * Use the period’s start time as `week_start` for plotting.
  * Draw a `sns.barplot` with formatted x-axis ticks every few weeks.

* **Why**

  * Weekly aggregation smooths daily noise and reveals seasonal patterns (e.g. higher volumes in spring/summer).
  * It provides a quick visual overview of how traffic evolves throughout the year.

---

### Visualisation: scatter plots with linear trend lines

* **What**
  Scatter plots of `daily_total` vs each meteorological variable (`temp_mean_C`, `precip_mm`, `wind_mean_ms`) with a simple linear trend line.

* **How**

  * For each selected column:

    * Drop rows with `NaN`.
    * Fit a 1D polynomial (degree 1) using `np.polyfit`.
    * Plot data points with `sns.scatterplot` (`alpha=0.4`) and overlay a red line based on the fitted polynomial.

* **Why**

  * Scatter plots show the cloud of daily observations and reveal the direction and strength of any global linear relationship.
  * The linear trend line is a simple, interpretable summary; it complements the grouped analyses and the correlation matrix.

---

### Extra analysis: grouped summaries for research questions

* **What**
  Grouped analyses to directly address the research questions:

  * Temperature bins vs mean daily counts.
  * Precipitation categories vs mean daily counts.
  * Weekday and weekend patterns.

* **How**

  * **Temperature bins**: `pd.cut` on `temp_mean_C` into custom ranges (`<0`, `0–10`, `10–15`, `15–20`, `20–25`, `25–30`, `>30`) and compute mean `daily_total` per bin.
  * **Precipitation categories**: `pd.cut` on `precip_mm` into `Dry`, `Light rain`, `Rainy` and compute mean `daily_total`.
  * **Weekday pattern**: group by `weekday`, compute mean `daily_total`, and enforce a logical weekday order.
  * **Weekday vs weekend**: group by `is_weekend` and map to labels `Weekday` / `Weekend`.
  * All summaries are printed and visualised via `sns.barplot`.

* **Why**

  * These grouped views directly target the core research questions:

    * Do moderate temperatures produce higher counts?
    * Does rain reduce traffic?
    * Are weekday patterns different from weekends?
  * Using bins and categories makes the non-linear nature of weather effects more explicit than a simple linear correlation.

---

### Correlation analysis

* **What**
  Compute and visualise the Pearson correlation matrix between `daily_total` and the main numeric weather features, and list absolute correlations with the target.

* **How**

  * Select numeric columns: `["daily_total", "temp_mean_C", "precip_mm", "wind_mean_ms"]`.
  * Compute `corr_df = df[numeric_cols].corr(method="pearson")`.
  * Plot a triangular heatmap with `sns.heatmap` (mask upper triangle, nice color scale).
  * Extract absolute correlations with `daily_total` and sort them descending.

* **Why**

  * The correlation matrix gives a compact, quantitative summary of linear relationships.
  * Even though correlations with `daily_total` are small, that insight is itself meaningful: it suggests that weather is only one of many drivers, and that the relationship is not purely linear, which is consistent with the bin-based analysis.

---

### Full pipeline and export

* **What**
  `run_pipeline()` orchestrates the full end-to-end process and writes the final merged dataset to CSV.

* **How**

  * Call all previous building blocks in order:

    1. `load_meteo_data`
    2. `aggregate_daily_meteo`
    3. `aggregate_velo_streaming` (with second aggregation)
    4. `check_missing_days` for both datasets
    5. `merge_datasets`
    6. `plot_weekly_bar`
    7. `plot_scatter_trends`
    8. `analyze_relationships`
    9. `analyze_correlations`
  * Save `merged` as `merged_dataset.csv`.

* **Why**

  * A single entry point makes the whole workflow reproducible and easy to re-run from raw CSVs.
  * This clearly demonstrates the “pipeline thinking” required by the learning outcomes: acquisition → cleaning → transformation → joining → analysis → export.


## 2) Choices Justification

* **Streaming Velo/Fuss with chunks**

  > “The Velo/Fuss CSV can be large, so we read it in chunks. This avoids loading the full file into memory and makes the pipeline scalable to more years or more locations.”

* **Double aggregation for Velo/Fuss**

  > “Chunked processing can split records for the same site and day across different chunks. To fix that, we first aggregate within each chunk and then apply a global groupby by (SiteCode, SiteName, date). This guarantees exactly one row per site-day in the final dataset.”

* **Different encodings (latin1 vs UTF-8)**

  > “MeteoSwiss exposes data in latin1, whereas the Basel portal uses UTF-8. we handle both encodings explicitly to avoid decoding errors and to preserve German characters in the site names.”

* **Date parsing and basic cleaning**

  > “I parse dates with a known format and use `errors='coerce'` plus `dropna` to remove invalid timestamps. For traffic counts, we coerce to numeric, drop non-numeric entries, and filter out negative values, which are not meaningful for volume data.”

* **Hourly to daily aggregation for MeteoSwiss**

  > “Weather data is originally hourly. For this project, we aggregate to daily summaries—mean temperature, total precipitation, and mean wind speed—so that its temporal resolution matches the daily traffic data.”

* **Checking missing days**

  > “I explicitly check for missing calendar days per year by comparing the present dates to a full date range. In 2024 both datasets have zero missing days, which simplifies the join and provides confidence in the coverage.”

* **Left join with traffic as the main table**

  > “Traffic counts are the primary object of interest, and weather provides context. Using a left join on `date` ensures that we keep all traffic days even if, hypothetically, weather information were missing for some dates.”

* **Calendar features: weekday and weekend flag**

  > “Day-of-week effects and weekend vs weekday patterns are key for mobility. we create `weekday` and `is_weekend` to separate commuting behavior from weekend behavior, which is necessary to interpret the weather effects correctly.”

* **Temperature bins and precipitation categories**

  > “Weather effects on traffic are unlikely to be purely linear. we use temperature bins and precipitation categories to capture non-linear patterns, such as a plateau around moderate temperatures and sharp drops on rainy days.”

* **Scatter plots with linear trend lines**

  > “Scatter plots with a linear trend line give a quick visual check for global linear relationships between traffic and each weather variable. They complement the grouped analyses and the correlation matrix.”

* **Correlation matrix**

  > “The Pearson correlation matrix summarises linear associations between daily totals and weather variables. The small coefficients indicate that weather explains only a modest share of the variability, which is consistent with the grouped results and the strong calendar effects.”

* **Single entry point: `run_pipeline()`**
